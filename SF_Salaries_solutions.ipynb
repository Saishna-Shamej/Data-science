```python
def CHIEF(JobTitle):
  if 'CHIEF' in JobTitle.upper().split():
    return True
  else:
    return False
sal = pd.read_csv('Salaries.csv')
sum(sal['JobTitle'].apply(lambda x: CHIEF(x)))
```
