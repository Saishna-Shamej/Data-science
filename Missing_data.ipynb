```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,6,np.nan],
                  'C':[1,2,3],
                  'D':[2,3,6],
                  'E':[6, 7, 8],
                  'F':[np.nan,1,np.nan]})
```

```python
df
```

dropna() removes the rows that contain null values.

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(thresh=1)
```

fillna() replaces the null values with a specified value.

```python
df.fillna(value="Fill value")
```

```python
df['A'].fillna(value=df['A'].mean())
```
